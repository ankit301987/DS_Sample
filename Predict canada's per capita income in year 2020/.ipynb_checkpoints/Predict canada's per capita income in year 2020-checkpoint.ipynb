<h3 style="color:purple">Sample problem of Canada per capita income</h3>

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df['per capita income (US$)'],color='red',marker='+')
```

```python
newdf=df.drop('per capita income (US$)',axis='columns')
newdf
```

```python
reg = linear_model.LinearRegression()
reg.fit(newdf,df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y=range(1970,2024)
y
```

```python
x=pd.Series(y)
x
```

```python
y=reg.coef_*x+reg.intercept_
y
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.plot(x,y,color='red')
plt.scatter(df.year,df['per capita income (US$)'],color='blue',marker='+')
```
