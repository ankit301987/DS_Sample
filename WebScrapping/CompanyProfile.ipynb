```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
req = Request('https://in.investing.com/equities/itc-company-profile'
                , headers={'User-Agent': 'Mozilla/5.0'})
```

```python
pages = urlopen(req).read()
```

```python
soup = BeautifulSoup(pages.decode('utf-8'),'html.parser')
```

```python
list_value = list(soup.children)
```

```python
html_page = list_value[3]
```

```python
html_body=list(html_page.children)[3]
```

```python
html_body.findAllNext('h2',class_='section-title')[0].get_text().strip()
```

```python
dic = dict()
```

```python
dic['Company'] = html_body.findAllNext('h2',class_='section-title')[0].get_text().strip()
```

```python
profile_details = html_body.findAllNext('dl',class_='profile-details-list')[0]
anchor = profile_details.findAllNext('a',class_='common-link')
```

```python
 dic['Industry'] = anchor[0].get_text()         
```

```python
 dic['Sector'] = anchor[1].get_text()         
```

```python
 dic['Employees'] =profile_details.findAllNext('dd',class_='profile-details-details')[2].get_text().strip()
```

```python
 dic['Equity Type'] = profile_details.findAllNext('dd',class_='profile-details-details')[3].get_text().strip()
```

```python
dic
```

```python
brief_details = html_body.findAllNext('span', class_='text common-pre-text')[0].get_text()
```

```python
dic['Company Details'] = brief_details
```

```python
dic
```

```python
table = html_body.findAllNext('table', class_='common-table e-executive')
```

```python
rows = table[0].findAll('tbody')[0].findAll('tr')
```

```python
rows
```

```python
rows[0].findAll('span')
```

```python
list_collection = list()
for row in rows:
    span_list = row.findAll('span')
    dic_executive = dict()
    dic_executive['Name']=span_list[0].get_text()
    dic_executive['Age']=span_list[1].get_text()
    dic_executive['Since']=span_list[2].get_text()
    dic_executive['Title']=span_list[3].get_text()
    list_collection.append(dic_executive)
    dic_executive = None
```

```python
list_collection
```

```python
import pandas as pd
data_frame = pd.DataFrame([dic])
```

```python
data_frame
```

```python
top_executive = pd.DataFrame(list_collection)
```

```python
top_executive
```

```python
final_data = pd.concat([data_frame,top_executive ], axis=1)
```

```python
final_data
```

```python
final_data.to_csv('ITC_CompanyProfile.csv')
```
