```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
req = Request('https://in.investing.com/equities/itc-historical-data-earnings'
                , headers={'User-Agent': 'Mozilla/5.0'})
```

```python
pages = urlopen(req).read()
```

```python
soup = BeautifulSoup(pages.decode('utf-8'),'html.parser')
```

```python
list_value = list(soup.children)
```

```python
html_page = list_value[3]
```

```python
html_body=list(html_page.children)[3]
```

```python
table_intended =  html_body.findAllNext('table')[1]
```

```python
tbody = list(table_intended.children)[5]
```

```python
rows = tbody.findAll('tr')
```

```python
rows[1].findAll('span')
```

```python
list_collection = list()
for row in rows:
    span_list = row.findAll('span')
    dic = dict()
    dic['Release Date']=span_list[0].get_text()
    dic['Period End']=span_list[1].get_text()
    dic['EPS Actual']=span_list[2].get_text()
    dic['EPS Forecast']=span_list[3].get_text()
    dic['EPS Surprise']=span_list[4].get_text()
    dic['Revenue Actual']=span_list[5].get_text()
    dic['Revenue Forecast']=span_list[4].get_text()
    dic['Revenue Surprise']=span_list[5].get_text()
    list_collection.append(dic)
    dic = None
```

```python
import pandas as pd
data_frame = pd.DataFrame(list_collection)
data_frame.head(15)
```

```python
date_inex = pd.to_datetime(data_frame['Release Date'])
data_frame['Release Date'] = date_inex
data_frame
```

```python
data_frame.to_csv('ITC_Earnings.csv')
```
