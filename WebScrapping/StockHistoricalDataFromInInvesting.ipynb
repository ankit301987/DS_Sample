```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
req = Request('https://in.investing.com/equities/itc-historical-data?end_date=1581532200&st_date=1388514600'
                , headers={'User-Agent': 'Mozilla/5.0'})
```

```python
pages = urlopen(req).read()
```

```python
soup = BeautifulSoup(pages.decode('utf-8'),'html.parser')
```

```python
list_value = list(soup.children)
```

```python
html_page = list_value[3]
```

```python
html_body=list(html_page.children)[3]
```

```python
table_intended =  html_body.findAllNext('table')[1]
```

```python
tbody = list(table_intended.children)[5]
```

```python
rows = tbody.findAll('tr')
```

```python
list_collection = list()
for row in rows:
    span_list = row.findAll('span')
    dic = dict()
    dic['Date']=span_list[0].get_text()
    if(len(span_list)==7):
        dic['Last Close Price']=span_list[1].get_text()
        dic['Open']=span_list[2].get_text()
        dic['High']=span_list[3].get_text()
        dic['Low']=span_list[4].get_text()
        dic['Volume']=span_list[5].get_text()
        dic['Percentage Changes']=span_list[6].get_text()
        dic['Events'] = ""
    else:
        dic['Last Close Price']=span_list[2].get_text()
        dic['Open']=span_list[3].get_text()
        dic['High']=span_list[4].get_text()
        dic['Low']=span_list[5].get_text()
        dic['Volume']=span_list[6].get_text()
        dic['Percentage Changes']=span_list[7].get_text()
        dic['Events'] = span_list[1].get_text()
    list_collection.append(dic)
    dic = None
    
```

```python
import pandas as pd
data_frame = pd.DataFrame(list_collection)
data_frame.head(15)
```

```python
result = data_frame[data_frame['Last Close Price'] == 'E']
result
```

```python
for i in list(result.index):
    data_frame.at[i,"Last Close Price"]= data_frame.at[i,"Open"]
    data_frame.at[i,"Open"] = data_frame.at[i,"High"]
    data_frame.at[i,"High"] = data_frame.at[i,"Low"]
    data_frame.at[i,"Low"] = data_frame.at[i,"Volume"]
    data_frame.at[i,"Volume"] = data_frame.at[i,"Percentage Changes"]
    percentage = 100*(float(data_frame.at[i,"Last Close Price"]) - float(data_frame.at[i,"Open"])) / float(data_frame.at[i,"Open"])
    data_frame.at[i,"Percentage Changes"] = '{0:.2f}'.format(percentage) + '%'
```

```python

def ConvertingVolumeInFloat(volume):
    if(volume[-1] == 'M'):
        return float(volume[:len(volume)-1])* 1000000
    elif(volume[-1] == 'K'):
        return float(volume[:len(volume)-1])* 1000
    else:
        return float(volume)
```

```python
result = data_frame.Volume.apply(lambda x: ConvertingVolumeInFloat(x))
result
```
